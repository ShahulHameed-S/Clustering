```python
a=d.iloc[:,3:].values
a
```

```python
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster import hierarchy as sch
```

```python
sch.dendrogram(sch.linkage(a, method='ward'))
plt.show()
```

```python
model=AgglomerativeClustering(n_clusters=5, linkage='ward',metric="euclidean")
label=model.fit_predict(a)

plt.scatter(a[label==0, 0], a[label==0, 1], c='green', s=100, label='c1')
plt.scatter(a[label==1, 0], a[label==1, 1], c='red', s=100, label='c2')
plt.scatter(a[label==2, 0], a[label==2, 1], c='blue', s=100, label='c3')
plt.scatter(a[label==3, 0], a[label==3, 1], c='yellow', s=100, label='c4')
plt.scatter(a[label==4, 0], a[label==4, 1], c='orange', s=100, label='c5')
centroids = []
for i in range(5):
    centroids.append(a[model == i].mean(axis=0))
centroids = np.array(centroids)
plt.scatter(centroids[:, 0],centroids[:, 1],color='red',label='Centroids',s=300,marker='>')
plt.show()
```

```python
label
```

```python
from sklearn.metrics import silhouette_score

score = silhouette_score(a, label)

print("Silhouette Score:", score)
```

```python
#Divisive

import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

n_c=[a]
n_cluster=5

while len(n_c)<n_cluster:
  model=KMeans(n_clusters=2)
  larg_cluster=max(n_c, key=len)
  # n_c.pop(n_c.index(larg_cluster))
  largest_index =  np.argmax([len(c) for c in n_c])
  n_c.pop(largest_index)
  labels=model.fit_predict(larg_cluster)


  cluster_1=larg_cluster[labels==0]
  cluster_2=larg_cluster[labels==1]

  n_c.append(cluster_1)
  n_c.append(cluster_2)

print(len(n_c))

colors = ['red', 'blue', 'green', 'orange', 'black']
for i, cluster in enumerate(n_c):
  plt.scatter(cluster[:,0], cluster[:,1],c=colors[i],label=f"cluster {i+1}", s=100)
plt.show()
```
