```python
transaction=[]
for i in range(len(d)):
  transaction.append([str(d.values[i,j])for j in range (20)if str(d.values[i,j])!="nan"])
```

```python
from apyori import apriori
ap=apriori(transactions=transaction,min_confidence=0.2,min_support=0.003,min_lenght=2,max_length=2,min_lift=3)
result=list(ap)
result
```

```python
import pandas as pd

# Create empty lists to store the rules data
lhss, rhss, supports, confidences, lifts = [], [], [], [], []

# Loop through the results to extract information
for record in result:
    support = record.support
    for ordered_stat in record.ordered_statistics:
        # Extract Antecedent (Left Hand Side) and Consequent (Right Hand Side)
        lhs = tuple(ordered_stat.items_base)
        rhs = tuple(ordered_stat.items_add)
        confidence = ordered_stat.confidence
        lift = ordered_stat.lift

        # Append data to lists
        lhss.append(', '.join(lhs))
        rhss.append(', '.join(rhs))
        supports.append(support)
        confidences.append(confidence)
        lifts.append(lift)

# Create the pandas DataFrame
df_results = pd.DataFrame({
    'Left Hand Side': lhss,
    'Right Hand Side': rhss,
    'Support': supports,
    'Confidence': confidences,
    'Lift': lifts
})

# Sort the table by Lift to see the strongest relationships first
df_results = df_results.sort_values(by='Lift', ascending=False)

# Display the table
print(df_results)
```

```python
import pandas as pd

# Step 1: Create an empty list to store extracted values
data = []

# Step 2: Loop through each RelationRecord in result
for record in result:

    # Step 3: Each record may contain multiple rules (usually 1 here)
    for stat in record.ordered_statistics:

        # Step 4: Extract required values and store in list
        base = list(stat.items_base)        # Left side (IF)
        add = list(stat.items_add)          # Right side (THEN)
        support = record.support            # Support of rule
        confidence = stat.confidence        # Confidence
        lift = stat.lift                    # Lift

        # Step 5: Append everything as a row
        data.append([base, add, support, confidence, lift])

# Step 6: Convert list into DataFrame (table)
df = pd.DataFrame(data, columns=[
    'Base Item',
    'Add Item',
    'Support',
    'Confidence',
    'Lift'
])


# Step 8: Show final table
print(df)
```
